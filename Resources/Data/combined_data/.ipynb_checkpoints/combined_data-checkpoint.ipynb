```python
import pandas as pd 
```

```python
data2018 = pd.read_csv("combined_2018.csv") 
data2017 = pd.read_csv("combined_2017.csv")
data2016 = pd.read_csv("combined_2016.csv")
```

```python
movies = pd.concat([data2018, data2017, data2016], ignore_index=True, sort =False)
```

```python
movies
```

```python
movies.to_csv('movies.csv', index=False)
```
