```python
import pandas as pd 
```

```python
budge_df = pd.read_csv("Resources/Data/movies_list.csv") 
```

```python
youtube_df = pd.read_csv("Resources/Data/youtube_2018.csv")
```

```python
budge_df
```

```python
youtube_df
```

```python
# Define a function to parse the money
def correct_name(row):
    name = row.replace(" Official Trailer", "")
    return name
```

```python
youtube_df['movie_trailer'] = youtube_df['movie_trailer'].apply(correct_name)
```

```python
youtube_df
```

```python
result = pd.merge(budge_df, youtube_df, left_on=['Movie'], right_on=['movie_trailer'],how='inner')
```

```python
result
```

```python
movies = result[['Movie', 'Production Budget', 'Domestic Gross', 'comment_count', 'view_count', 'like_count', ]]

```

```python
movies.rename(columns = {'Movie':'movie', 'Production Budget':'budge', 'Domestic Gross':'gross'}, inplace = True)

```

```python
movies
```

```python
movies.to_csv('Resources/Data/combined_2018.csv', index=False)
```
